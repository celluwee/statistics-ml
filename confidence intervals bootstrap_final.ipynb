**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

Для ускорения расчетов можете использовать broadcasting (не является обязательным требованием). Учите, что в таком случае внутри функции у вас могут получиться очень большие матрицы, из-за которых может кончиться оперативная память. 

```python
def bootstrap(sample, B=100000): 
    '''
    Считает бутстрепные оценки для исходной выборки

    :param sample: исходная выборка
    :param B: количество бутстрепных 
    :return bootstrap_estimations: оценки по бутстрепным выборкам
    '''   
    n = len(sample)
    bootstrap_estimations = np.zeros(B)
    
    for i in range(B):
        bootstrap_sample = np.random.choice(sample, size=n, replace=True)
        bootstrap_estimations[i] = np.mean(bootstrap_sample)
    
    return bootstrap_estimations

    raise NotImplementedError
```

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу нормального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    mean_bootstrap = np.mean(bootstrap_estimations)
    std_err_bootstrap = np.std(bootstrap_estimations)
    z_score = np.abs(stats.norm.ppf((1 - alpha) / 2))
    left = mean_bootstrap - z_score * std_err_bootstrap
    right = mean_bootstrap + z_score * std_err_bootstrap
    
    return left, right

    raise NotImplementedError

    
def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу центрального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    mean_bootstrap = np.mean(bootstrap_estimations)
    left = 2 * mean_bootstrap - np.quantile(bootstrap_estimations, 1 - ((1 - alpha) / 2))
    right = 2 * mean_bootstrap - np.quantile(bootstrap_estimations, (1 - alpha) / 2)
    
    raise NotImplementedError

    
def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу квантильного бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    quantile_left = (1 - alpha) / 2
    quantile_right = 1 - quantile_left
    left = np.quantile(bootstrap_estimations, quantile_left)
    right = np.quantile(bootstrap_estimations, quantile_right)

    return left, right

    raise NotImplementedError
```

**Решение:**

```python
sample = sps.cauchy.rvs(size=1000)
len(sample)
```

```python
from scipy.stats import trim_mean

cum_mean = []
cum_trim_mean = []
cum_med = []

for n in range(1, 1001):
    cum_mean.append(np.mean(sample[:n]))
    cum_trim_mean.append(trim_mean(sample[:n], 0.1))  # примем альфа за 0.1, считать тяжело
    cum_med.append(np.nanquantile(sample[:n], 0.5))    
    
plt.figure(figsize=(15, 10))

plt.plot(cum_mean, label="среднее")
plt.plot(cum_trim_mean, label="усеченное среднее")
plt.plot(cum_med, label="медиана")

plt.xlabel='n'
plt.title("зависимость значений всех оценок от n для одной реализации выборки")
plt.legend()

plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mean_error = []
trim_mean_error = [] 
med_error = []

for n in range(1, 1001):
    mean_error.append(mean_squared_error(cum_mean[:n], np.zeros(n)))
    trim_mean_error.append(mean_squared_error(cum_trim_mean[:n], np.zeros(n))) 
    med_error.append(mean_squared_error(cum_med[:n], np.zeros(n))) 
    
plt.figure(figsize=(15, 10))

plt.plot(mean_error, label="mse для среднего")
plt.plot(trim_mean_error, label="mse для усеченного среднего")
plt.plot(med_error, label="mse для медианы")

plt.legend()

plt.show()
```

**Вывод:** по свойству медианы и усеченного среднего они асимптотически стремятся при росте n к значению оцениваемого параметра (0) для Коши, среднее является смещенной оценкой, для медианы получаем наименьшую среднеквадратичную ошибку, наибольшую - для среднего.
